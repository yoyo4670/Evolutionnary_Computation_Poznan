```python
#Read the point coordonnates from the file TSPA.csv and plot those points

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

#Read the file
df = pd.read_csv('./src/TSPD.csv', sep=';',header=None)
df.columns = ['x','y','weight']

#Plot the points with the weight as the color of the point (the bigger the weight, the clearer the point)
plt.scatter(df['x'],df['y'],c=df['weight'],cmap='viridis')
plt.colorbar()
plt.show()


```

```python
#The greedyCycle.csv file contains the index of the points inside TSPA in the order they are visited
#Read the file
dfGreedy = pd.read_csv('./greedyLocalSearchBestStartNodesExchange.csv', sep=';',header=None)
dfGreedy.columns = ['index']
#Link the index to the coordonnates
dfGreedy = dfGreedy.join(df, on='index', how='left')
#Add the first point at the end to close the cycle
dfGreedy.loc[len(dfGreedy)] = dfGreedy.loc[0]
#Plot the lines between the points and the points with the weight as the color of the point (the bigger the weight, the clearer the point)
plt.plot(dfGreedy['x'],dfGreedy['y'])
plt.scatter(df['x'],df['y'],c=df['weight'],cmap='viridis',s=50)
plt.colorbar()
plt.show()
```
